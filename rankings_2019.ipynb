```python
import pandas as pd
import numpy as np
```

```python
df_adp_ev = pd.read_csv('data/values.csv')
df_ranks_raw = pd.read_csv('data/ffa_customrankings2019-0.csv')
df_ranks_raw = df_ranks_raw[['player','position','points','lower','upper','sdPts','positionRank','overallECR',
                            'positionECR','risk','sleeper','adp']]
```

```python
df_adp_ev.columns = ['positionRank','position','ev']
```

```python
df = df_adp_ev.merge(df_ranks_raw, how='inner', on=['position','positionRank'])
df.sort_values(by=['ev','positionRank'], ascending=[False,True]).reset_index(drop=True).head(50)
```
